## <font color='#9BFDFF'>Criterios y justificaci√≥n</font>




Para este an√°lisis, se seleccionaron cuatro criterios fundamentales que se consideran indicadores clave del √©xito cinematogr√°fico:<br><br>

**1) Puntuaci√≥n en IMDb (‚â• 7.0):**

Justificaci√≥n: IMDb es una de las plataformas m√°s influyentes para la opini√≥n
del p√∫blico. Las pel√≠culas con una puntuaci√≥n igual o superior a 7.0 suelen ser vistas como de alta calidad, ya que reflejan una valoraci√≥n positiva por parte de una audiencia amplia. Un estudio de Wallace et al. (2016) sugiere que las pel√≠culas con alta calificaci√≥n en IMDb tienden a mantenerse relevantes a largo plazo, pues la valoraci√≥n del p√∫blico influye directamente en su permanencia en la memoria colectiva.

**Fuente:** Wallace, W., Seigerman, A., & Holbrook, M. (2016). The Role of Critics in the Culture of the Movie Industry. Journal of Marketing, 61(2), 42‚Äì55. <br><br>

**2) Ingresos Brutos (superiores a la mediana del dataset):**

Justificaci√≥n: Los ingresos brutos son un indicador directo de la rentabilidad financiera de una pel√≠cula. Sin embargo, los ingresos deben analizarse en relaci√≥n con el presupuesto de producci√≥n para obtener una medida precisa de √©xito financiero. De Vany y Walls (1999) argumentan que el retorno de inversi√≥n es un factor cr√≠tico, pero debe ser considerado dentro de un contexto que contemple la incertidumbre inherente a la industria cinematogr√°fica.

**Fuente:** De Vany, A., & Walls, W. D. (1999). Uncertainty in the Movie Industry: Does Star Power Reduce the Terror of the Box Office? Journal of Cultural Economics, 23(4), 285‚Äì318.
<br><br>

**3) Votos de Usuarios (m√°s de 100,000 votos):**

Justificaci√≥n: El n√∫mero de votos y rese√±as en plataformas como IMDb refleja la popularidad y aceptaci√≥n de una pel√≠cula entre el p√∫blico general. Las pel√≠culas que generan una mayor interacci√≥n y participaci√≥n por parte de los usuarios tienden a ser m√°s exitosas en t√©rminos de popularidad y tienen un impacto m√°s duradero. Mishra y Rout (2020) destacan que las m√©tricas sociales y de engagement digital permiten una mejor estimaci√≥n del √©xito de una pel√≠cula desde la perspectiva del p√∫blico general.

**Fuente:** Mishra, A., & Rout, H. K. (2020). Social Media and the Film Industry: Measuring Popularity and Success. International Journal of Communication, 14, 1365‚Äì1383.
<br><br>

**4) Reconocimiento Cr√≠tico (premios y distinciones):**

Justificaci√≥n: El reconocimiento cr√≠tico es esencial para evaluar la importancia art√≠stica de una pel√≠cula. Las pel√≠culas que reciben elogios de la cr√≠tica y ganan premios importantes, como los √ìscar o los Globos de Oro, tienen un valor cultural significativo que las distingue de otras pel√≠culas. Konnikova (2014) menciona que algunas pel√≠culas, como "Frozen", logran un impacto cultural profundo m√°s all√° de su rendimiento financiero, gracias a su resonancia emocional y narrativa.

**Fuente:** Konnikova, M. (2014). How ‚ÄòFrozen‚Äô Took Over the World. The New Yorker. https://www.newyorker.com

## <font color='#5BFFB1'>Gr√°fico 3: Duraci√≥n de la Pel√≠cula vs. Puntuaci√≥n en IMDb</font>

Identificar si las pel√≠culas largas o cortas tienden a obtener mejores o peores calificaciones y revisar si existen posibles valores at√≠picos.

```python
# Box Plot

# Agrupar duraci√≥n en categor√≠as
bins = [0, 80, 100, 120, 140, 300]
labels = ['<80 min', '80-100 min', '100-120 min', '120-140 min', '>140 min']
df['duration_group'] = pd.cut(df['duration'], bins=bins, labels=labels)

# Crear el box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='duration_group', y='imdb_score', data=df, palette='Set1')

plt.title('Duraci√≥n de la Pel√≠cula vs. Puntuaci√≥n en IMDb')
plt.xlabel('Duraci√≥n de la Pel√≠cula')
plt.ylabel('Puntuaci√≥n IMDb')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## <font color='#5BFFB1'>Gr√°fico 4: N√∫mero de Votos de Usuarios por G√©nero</font>

Ver qu√© g√©neros generan mayor interacci√≥n del p√∫blico, lo que puede ser un buen indicativo de la popularidad.

```python
# Gr√°fico de barras

# Expandir los g√©neros a filas individuales
df_exploded = df.copy()
df_exploded['genres'] = df_exploded['genres'].str.split('|')
df_exploded = df_exploded.explode('genres')

# Agrupar por g√©nero individual
votos_por_genero_individual = df_exploded.groupby('genres')['num_voted_users'].sum().sort_values(ascending=False).head(10)

# Graficar nuevamente
plt.figure(figsize=(12, 6))
sns.barplot(x=votos_por_genero_individual.index, y=votos_por_genero_individual.values, palette='viridis')
plt.title('N√∫mero de Votos de Usuarios por G√©nero Individual (Top 10)')
plt.xlabel('G√©nero')
plt.ylabel('N√∫mero de Votos de Usuarios')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## <font color='#5BFFB1'>Gr√°fico 5: Distribuci√≥n de Ingresos Brutos por G√©nero</font>

 Identificar qu√© g√©neros tienden a generar m√°s ingresos y la dispersi√≥n dentro de cada uno.

```python
# Violin Plot


# Limpiar y tomar solo el primer g√©nero
df_clean = df[['gross', 'genres']].dropna()
df_clean['main_genre'] = df_clean['genres'].apply(lambda x: x.split('|')[0])

# Seleccionar los 10 g√©neros m√°s frecuentes
top_genres = df_clean['main_genre'].value_counts().nlargest(10).index
df_top = df_clean[df_clean['main_genre'].isin(top_genres)]

# Graficar
plt.figure(figsize=(12, 6))
sns.violinplot(x='main_genre', y='gross', data=df_top, inner='box', palette='Set2')
plt.title('Distribuci√≥n de Ingresos Brutos por G√©nero Principal')
plt.xlabel('G√©nero')
plt.ylabel('Ingresos Brutos (gross)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## <font color='#5BFFB1'>Gr√°fico 8: Comparaci√≥n de Pel√≠culas con √âxito Cr√≠tico vs √âxito de Taquilla</font>

Permitire una visualizaci√≥n clara de c√≥mo se distribuyen las pel√≠culas exitosas en ambas dimensiones.

```python
# Gr√°fico de barras apiladas

# Crear dos categor√≠as para el √©xito cr√≠tico y de taquilla
# Definir umbrales para determinar si una pel√≠cula es exitosa en cada dimensi√≥n
umbral_critico = df['num_critic_for_reviews'].median()  # Media de cr√≠ticas
umbral_taquilla = df['gross'].median()  # Media de ingresos de taquilla

# Crear una nueva columna para clasificar el √©xito cr√≠tico
df['exito_critico'] = df['num_critic_for_reviews'].apply(lambda x: 'Exitoso' if x >= umbral_critico else 'No exitoso')

# Crear una nueva columna para clasificar el √©xito en taquilla
df['exito_taquilla'] = df['gross'].apply(lambda x: 'Exitoso en Taquilla' if x >= umbral_taquilla else 'No exitoso en Taquilla')

# Contar el n√∫mero de pel√≠culas en cada categor√≠a
exito_combinado = pd.crosstab(df['exito_critico'], df['exito_taquilla'])

# Crear gr√°fico de barras apiladas
exito_combinado.plot(kind='bar', stacked=True, color=['#4CAF50', '#FF0000'], figsize=(10,6))

# Configuraci√≥n del gr√°fico
plt.title('Comparaci√≥n de Pel√≠culas con √âxito Cr√≠tico vs √âxito en Taquilla')
plt.xlabel('√âxito Cr√≠tico')
plt.ylabel('N√∫mero de Pel√≠culas')
plt.xticks(rotation=0)
plt.legend(title="√âxito en Taquilla", loc='upper left')

# Mostrar gr√°fico
plt.tight_layout()
plt.show()
```
