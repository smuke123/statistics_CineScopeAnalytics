## <font color='#9BFDFF'>Criterios y justificaci√≥n</font>




Para este an√°lisis, se seleccionaron cuatro criterios fundamentales que se consideran indicadores clave del √©xito cinematogr√°fico:<br><br>

**1) Puntuaci√≥n en IMDb (‚â• 7.0):**

Justificaci√≥n: IMDb es una de las plataformas m√°s influyentes para la opini√≥n
del p√∫blico. Las pel√≠culas con una puntuaci√≥n igual o superior a 7.0 suelen ser vistas como de alta calidad, ya que reflejan una valoraci√≥n positiva por parte de una audiencia amplia. Un estudio de Wallace et al. (2016) sugiere que las pel√≠culas con alta calificaci√≥n en IMDb tienden a mantenerse relevantes a largo plazo, pues la valoraci√≥n del p√∫blico influye directamente en su permanencia en la memoria colectiva.

**Fuente:** Wallace, W., Seigerman, A., & Holbrook, M. (2016). The Role of Critics in the Culture of the Movie Industry. Journal of Marketing, 61(2), 42‚Äì55. <br><br>

**2) Ingresos Brutos (superiores a la mediana del dataset):**

Justificaci√≥n: Los ingresos brutos son un indicador directo de la rentabilidad financiera de una pel√≠cula. Sin embargo, los ingresos deben analizarse en relaci√≥n con el presupuesto de producci√≥n para obtener una medida precisa de √©xito financiero. De Vany y Walls (1999) argumentan que el retorno de inversi√≥n es un factor cr√≠tico, pero debe ser considerado dentro de un contexto que contemple la incertidumbre inherente a la industria cinematogr√°fica.

**Fuente:** De Vany, A., & Walls, W. D. (1999). Uncertainty in the Movie Industry: Does Star Power Reduce the Terror of the Box Office? Journal of Cultural Economics, 23(4), 285‚Äì318.
<br><br>

**3) Votos de Usuarios (m√°s de 100,000 votos):**

Justificaci√≥n: El n√∫mero de votos y rese√±as en plataformas como IMDb refleja la popularidad y aceptaci√≥n de una pel√≠cula entre el p√∫blico general. Las pel√≠culas que generan una mayor interacci√≥n y participaci√≥n por parte de los usuarios tienden a ser m√°s exitosas en t√©rminos de popularidad y tienen un impacto m√°s duradero. Mishra y Rout (2020) destacan que las m√©tricas sociales y de engagement digital permiten una mejor estimaci√≥n del √©xito de una pel√≠cula desde la perspectiva del p√∫blico general.

**Fuente:** Mishra, A., & Rout, H. K. (2020). Social Media and the Film Industry: Measuring Popularity and Success. International Journal of Communication, 14, 1365‚Äì1383.
<br><br>

**4) Reconocimiento Cr√≠tico (premios y distinciones):**

Justificaci√≥n: El reconocimiento cr√≠tico es esencial para evaluar la importancia art√≠stica de una pel√≠cula. Las pel√≠culas que reciben elogios de la cr√≠tica y ganan premios importantes, como los √ìscar o los Globos de Oro, tienen un valor cultural significativo que las distingue de otras pel√≠culas. Konnikova (2014) menciona que algunas pel√≠culas, como "Frozen", logran un impacto cultural profundo m√°s all√° de su rendimiento financiero, gracias a su resonancia emocional y narrativa.

**Fuente:** Konnikova, M. (2014). How ‚ÄòFrozen‚Äô Took Over the World. The New Yorker. https://www.newyorker.com

## <font color='#5BFFB1'>Gr√°fico 8: Comparaci√≥n de Pel√≠culas con √âxito Cr√≠tico vs √âxito de Taquilla</font>

Permitire una visualizaci√≥n clara de c√≥mo se distribuyen las pel√≠culas exitosas en ambas dimensiones.

```python
# Gr√°fico de barras apiladas

# Crear dos categor√≠as para el √©xito cr√≠tico y de taquilla
# Definir umbrales para determinar si una pel√≠cula es exitosa en cada dimensi√≥n
umbral_critico = df['num_critic_for_reviews'].median()  # Media de cr√≠ticas
umbral_taquilla = df['gross'].median()  # Media de ingresos de taquilla

# Crear una nueva columna para clasificar el √©xito cr√≠tico
df['exito_critico'] = df['num_critic_for_reviews'].apply(lambda x: 'Exitoso' if x >= umbral_critico else 'No exitoso')

# Crear una nueva columna para clasificar el √©xito en taquilla
df['exito_taquilla'] = df['gross'].apply(lambda x: 'Exitoso en Taquilla' if x >= umbral_taquilla else 'No exitoso en Taquilla')

# Contar el n√∫mero de pel√≠culas en cada categor√≠a
exito_combinado = pd.crosstab(df['exito_critico'], df['exito_taquilla'])

# Crear gr√°fico de barras apiladas
exito_combinado.plot(kind='bar', stacked=True, color=['#4CAF50', '#FF0000'], figsize=(10,6))

# Configuraci√≥n del gr√°fico
plt.title('Comparaci√≥n de Pel√≠culas con √âxito Cr√≠tico vs √âxito en Taquilla')
plt.xlabel('√âxito Cr√≠tico')
plt.ylabel('N√∫mero de Pel√≠culas')
plt.xticks(rotation=0)
plt.legend(title="√âxito en Taquilla", loc='upper left')

# Mostrar gr√°fico
plt.tight_layout()
plt.show()
```

## <font color='#5BFFB1'>Gr√°fico 9: Teoremas celebres</font>

La **media poblacional** de una variable \( X \) se define como:

$$
\mu = \frac{1}{N} \sum_{i=1}^{N} x_i
$$
Donde:
- $ \mu $: es la media poblacional,
- $ x_i $: es el valor de cada observaci√≥n,
- $ N $: es el n√∫mero total de datos.

La **desviaci√≥n est√°ndar poblacional** se define como:

$$
\sigma = \sqrt{ \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2 }
$$
Donde:
- $\sigma$: es la desviaci√≥n est√°ndar poblacional,
- $\mu$: es la media,
- $x_i$: es cada valor observado,
- $N$: es el n√∫mero total de datos.

La **funci√≥n de densidad de probabilidad** de una distribuci√≥n normal es:

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \cdot e^{ -\frac{(x - \mu)^2}{2\sigma^2} }
$$
Donde:
- $ F(x) $: es la densidad de probabilidad para el valor \( x \),
- $ \mu $: es la media,
- $ \sigma $: es la desviaci√≥n est√°ndar.

Entonces en el dataset vamos a hallar la distribucion emp√≠rica obteniendo la media y desviaci√≥n estandar de los datos originales, graficamos el histograma y una curva de densidad.

```python
from scipy.stats import norm

data = df['gross'].dropna()

plt.figure(figsize=(10, 6))

sns.histplot(data, bins=30, kde=True, color='skyblue', edgecolor='black', stat='density', label='Datos originales')

# Opcional: A√±adir una curva normal te√≥rica para comparar
mean, std = data.mean(), data.std()
x = np.linspace(min(data), max(data), 100)
plt.plot(x, norm.pdf(x, mean, std), 'r--', label='Normal te√≥rica')

# Personalizaci√≥n
plt.title('Distribuci√≥n Original de la Variable "gross"', fontsize=16)
plt.xlabel('Ingresos brutos (gross)', fontsize=12)
plt.ylabel('Densidad', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

**Teorema del L√≠mite Central**

Este teorema establece que, si tomamos m√∫ltiples muestras aleatorias de tama√±o suficientemente grande ($n \geq 30$) de una poblaci√≥n con media $\mu$ y desviaci√≥n est√°ndar $\sigma$, la distribuci√≥n de las medias muestrales $\bar{X}$ tiende a una distribuci√≥n normal:

$$
\bar{X} \sim \mathcal{N}\left(\mu, \frac{\sigma^2}{n} \right)
$$

Esto significa que la media muestral seguir√° una distribuci√≥n normal, sin importar la distribuci√≥n original de los datos.


```python
from scipy.stats import norm

#Nueva variable: data
data = df['gross'].dropna()
sample_means = []

#Bucle para crear el array
for _ in range(1000):
    sample = np.random.choice(data, size=30, replace=True)
    sample_means.append(np.mean(sample))

mean_mu = np.mean(sample_means)
std_sigma = np.std(sample_means)

#Grafica
plt.figure(figsize=(10, 6))
sns.histplot(sample_means, bins=30, kde=False, stat='density', color='skyblue', edgecolor='black', label='Distribuci√≥n muestral')
x = np.linspace(min(sample_means), max(sample_means), 100)
plt.plot(x, norm.pdf(x, mean_mu, std_sigma), 'r--', label='Curva normal te√≥rica')
plt.title('Teorema del L√≠mite Central aplicado al gross')
plt.xlabel('Media de muestra')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```

**Independencia entre media y varianza muestrales**

Si los datos $X_1, X_2, \dots, X_n$ provienen de una distribuci√≥n normal, entonces la media muestral $\bar{X}$ y la varianza muestral $S^2$ son variables aleatorias **independientes**:

$$
S^2 = \frac{1}{n - 1} \sum_{i=1}^{n} (X_i - \bar{X})^2
$$

Este resultado solo se cumple si los datos provienen de una **distribuci√≥n normal**.

 El objetivo es observar si hay una dependencia entre ambas. Esto permite explorar la validez de una propiedad estad√≠stica clave. Si los puntos est√°n dispersos sin una tendencia clara, sugiere independencia.


```python
#Variables
medias = []
varianzas = []

#Tomamos 1000 muestras de 'gross' y a√±adimos al array
for _ in range(5000):
    muestra = np.random.choice(data, size=30, replace=True)
    medias.append(np.mean(muestra))
    varianzas.append(np.var(muestra, ddof=1))

#Grafico
plt.figure(figsize=(8,5))
plt.scatter(medias, varianzas, alpha=0.5)
plt.title('Relaci√≥n entre media y varianza muestral')
plt.xlabel('Media')
plt.ylabel('Varianza')
plt.grid(True)
plt.show()
```

**Ley de los grandes numeros**

Se genera una curva que muestra la media acumulada y se compara con la media real de todo el dataset. Esto demuestra la Ley de los Grandes N√∫meros, que afirma que la media muestral converge a la verdadera medida poblacional al aumentar el tama√±o de la muestra. El gr√°fico ilustra c√≥mo esa convergencia sucede progresivamente.

La ley establece que, al aumentar el n√∫mero de observaciones $n$, la media muestral $\bar{X}_n$ se aproxima a la media poblacional $\mu$:

$$
\lim_{n \to \infty} \bar{X}_n = \mu \quad \text{(con alta probabilidad)}
$$

Es decir, con suficientes datos, nuestras estimaciones se vuelven m√°s confiables.


```python
#Variables
media_real = np.mean(data)
medias_acumuladas = []

#A√±adimos el promedio de cada fila de 'gross'
for n in range(1, len(data)):
    medias_acumuladas.append(np.mean(data[:n]))

#Grafica
plt.figure(figsize=(10,5))
plt.plot(medias_acumuladas, label='Media muestral acumulada')
plt.axhline(media_real, color='red', linestyle='--', label='Media real')
plt.title('Ley de los Grandes N√∫meros')
plt.xlabel('Tama√±o de la muestra')
plt.ylabel('Media')
plt.legend()
plt.grid(True)
plt.show()
```
