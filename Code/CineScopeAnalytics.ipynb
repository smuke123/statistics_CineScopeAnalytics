<h1> Distribuci√≥n binomial </h1>

 En este an√°lisis se ha definido el "√©xito financiero" como lograr una ganancia de al menos 1.5 veces el presupuesto invertido. Usamos un modelo probabil√≠stico binomial para estimar la probabilidad de observar cierto n√∫mero de pel√≠culas exitosas en una muestra de pel√≠culas. Como la muestra es grande, usamos una aproximaci√≥n normal para representar mejor la distribuci√≥n esperada de √©xitos. Esto nos ayuda a visualizar c√≥mo se comporta el √©xito en el cine desde una perspectiva estad√≠stica.

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

# Consideramos el exito como la ganancia

df['exito_financiero'] = df['gross'] >= 1.5 * df['budget']

# Usamos las columnas sobrantes

grupo = df.dropna(subset=['budget', 'gross'])

n = len(grupo)
k = grupo['exito_financiero'].sum()
p_hat = k / n

print(f"Total de pel√≠culas: {n}")
print(f"√âxitos observados: {k}")
print(f"Probabilidad emp√≠rica de √©xito: {p_hat:.2f}")

# Hacemos el plot

x = range(n+1)
binom_dist = binom.pmf(x, n, p_hat)

plt.figure(figsize=(10, 5))
plt.bar(x, binom_dist, color='skyblue')
plt.axvline(k, color='red', linestyle='--', label=f'Observado: {k} √©xitos')
plt.title('Distribuci√≥n Binomial')
plt.xlabel('N√∫mero de √âxitos')
plt.ylabel('Probabilidad')
plt.legend()
plt.tight_layout()
plt.show()

# Graficamos una muestra

sample_df = df.dropna(subset=['budget', 'gross']).sample(30, random_state=1)
sample_df['exito_financiero'] = sample_df['gross'] >= 1.5 * sample_df['budget']

n = len(sample_df)
k = sample_df['exito_financiero'].sum()
p_hat = k / n
```
